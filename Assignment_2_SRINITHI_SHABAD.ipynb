# **Assignment 2 - Data Visualization**

## **Importing Necessary Libraries**

```python
# import libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Loading Car Crashes Dataset**

```python
df = sns.load_dataset("car_crashes")
df.head()
```

```python
df.info()
```

## **Data Visualization**

### **Distplots**

**Total Crashes**

```python
sns.distplot(df["total"], bins=5)
plt.show()
```

**Inference:** From the plot we can see that the total number of crashes across the US have a normal distribution.


---



**Insurance Premium**

```python
sns.distplot(df['ins_premium'], bins=5)
plt.show()
```

**Inference:** From the plot we can see that the distribution of the insurance paid by the drivers is skewed.



---



### **Bar Charts**

**Crashes per state**

```python
plt.subplots(figsize=(17, 5))
sns.barplot(data=df, x="abbrev", y="total")
plt.show()
```

Inference: From this bar chart we can identify states with the highest and lowest crash rates.



---



Crashes due to Alcohol Consumption per state

```python
plt.subplots(figsize=(17, 5))
sns.barplot(data=df, x="abbrev", y="alcohol")
plt.show()
```

Inference: From this bar chart we can identify states with the highest and lowest crash rates due to alcohol consumption.



---



### **Scatter Plots**

**First time crashes vs Total crashes**

```python
sns.scatterplot(data=df, y="total", x="no_previous", hue="abbrev")
plt.legend([],[], frameon=False)
plt.show()
```

**Inference:** From the plot we can see that the number of first time crashes is directly proportional to the total number of crashes, regardless of state.



---



**Insurance paid vs Crashes due to speeding**

```python
sns.scatterplot(data=df, y="ins_premium", x="speeding", hue="abbrev")
plt.legend([],[], frameon=False)
plt.show()
```

**Inference:** From the plot we can see that the insurance premium paid by drivers has no direct correlation with the amount of speed crashes



---



### **Joint Plots**

**Speeding vs Alcohol**

```python
sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg')
plt.show()
```

**Inference:** From this plot we can see that the speed crashes and alcohol crashes have skewed distributions but there is correlation as the speed crashes increase, alcohol crashes increase too, with some outliers.

### **Box Plots**

**Speeding Data**

```python
sns.boxplot(y='speeding', data=df)
plt.show()
```

**Inference:** From this box plot we can see the different quartiles of distribution as well as the median of the crashes caused due to speeding.

### **Pair Plots**


Total, Speeding, Alcohol

```python
sns.pairplot(df[['total', 'not_distracted', 'alcohol']])
plt.show()
```

**Inference:** From this pairplot we can see the relations between total crashes, crashes due to alcohol, and crashes where driver was not distracted. We can see a positive correlation between all 3.



---



### **Violin Plot**

**Speeding Data**

```python
sns.violinplot(y='speeding',data=df)
plt.show()
```

**Inference:** From this plot, we can see the distribution of the number of deaths caused by speeding. We notice that the highest number of instances of the deaths is a bit more than 4.



---



### **Swarm Plots**


**Alcohol Data**

```python
sns.swarmplot(y='alcohol' , data=df, hue='abbrev')
plt.legend([],[], frameon=False) # removing legend as it overflows
plt.show()
```

**Inference:** From the plot we can see more dots near 4, telling us that there is a higher number of instances for that value.



---



### **Strip Plots**

**Insurance Losses**

```python
sns.stripplot(y='ins_losses' , data=df, hue='abbrev')
plt.legend([],[], frameon=False) # removing legend as it overflows
plt.show()
```

**Inference:** From the plot we can see the different instances of the insurance losses, color coded by the state. We notice that most of the losses are around 140.



---



### **Heatmap**

```python
corr_matrix = df.iloc[:, :7] # ignoring abbrev field
sns.heatmap(corr_matrix.corr(), annot=True)
plt.show()
```

**Inference:** From this heatmap we can see how different fields are related to one another.
